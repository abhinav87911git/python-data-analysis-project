```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'C:\Users\abhin\Downloads\Python_Diwali_Sales_Analysis-main\Python_Diwali_Sales_Analysis-main\Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
# information about dataframe

df.info()
```

```python
# dropping the columns Status and unnamed1 because they have null values

df.drop(['Status','unnamed1'],  axis=1, inplace=True)
```

```python
df.info()
```

```python
#df.isnull().sum()
df.Amount.describe()
```

```python
# the column 'Amount' have 12 null values
# so making a column that contains values in True and False corresponding the 'Amount' column's null values

df['nulls'] = df.Amount.isnull()
```

```python
# making a new variable that contains only the rows in which Amount is null

onlynulls = df.loc[(df.nulls==True)]
```

```python
onlynulls
```

```python
# filling the value 00 in the place of null values in Amount column

df['Amount'].fillna(00, inplace=True)
```

```python
df.Amount.isnull().sum()
```

```python
# now the Amount column has zero null values

df.Amount.head(10)
```

```python
df.Amount = df.Amount.astype(int)
```

```python
df.Amount.describe()
```

#                       EXPLORATERY DATA ANALYSIS 

 ON THE BASIS OF GENDER

```python
df.Gender.value_counts()
```

```python
ax = sns.countplot(x= 'Gender', data = df, palette=['blue', 'orange'],hue = 'Gender',legend=False)
```

```python
ax = sns.countplot(x= 'Gender', data = df, palette=['blue', 'orange'],hue = 'Gender',legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum()
sales_gen
```

```python
sns.barplot(x='Gender', y='Amount', data= sales_gen,hue = 'Gender', width=0.8)
sns.set(rc={'figure.figsize':(7,5)})
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

   ANALYSIS ON THE BASIS OF AGE GROUP AND GENDER

```python
ax = sns.countplot(x ='Age Group', data= df, hue='Gender')
```

```python
ax = sns.countplot(x ='Age Group', data= df, hue='Gender', gap = 0.1)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount vs age group

sales_age = df.groupby('Age Group', as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
sales_age
```

```python
sns.barplot(x='Age Group', y ='Amount', data= sales_age, hue= 'Age Group')
```

FROM ABOVE GRAPHS WE CAN SEE THAT THE MORE BUYERS ARE FEMALES FROM 26-35 AGE GROUP

```python
df.columns
```

      STATE VISE ANALYSIS

```python
sns.set(rc={'figure.figsize': (10,10)})
ss= sns.countplot(y='State', data =df)
for bars in ss.containers:
    ss.bar_label(bars)
```

```python
#           total number of orders from top 10 states



sales_state = df.groupby('State', as_index= False)['Orders'].sum().sort_values(by='Orders', ascending= False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x='State', y= 'Orders', hue= 'State')
```

```python
#                total amount/sales from top 10 states

sales_state = df.groupby('State', as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)


sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x='State', y= 'Amount', hue= 'State')
```

WE CAN SEE THAT MOST ORDERS AND SALES ARE FROM UP, MAHARASTRA AND KARNATAKA STATES 

         ANALYSIS BASED ON MARITAL STATUS

```python
sns.set(rc={'figure.figsize':(8,5)})
ax = sns.countplot(data= df, x= 'Marital_Status', hue='Gender', width= 0.5)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital= df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_marital
```

```python
sns.barplot(data= sales_marital, x='Marital_Status', y= 'Amount', hue='Gender')

            
```

From above graphs we can see that most of  the buyers are unmarried (women) and they have high purchasing power*

 OCCUPATION BASED ANALYSIS

```python
ax = sns.countplot(data= df, x= 'Occupation', hue='Occupation', width= 0.5)
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation= df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.barplot(data= sales_occupation, x='Occupation', y= 'Amount', hue='Occupation')
```

FROM ABOVE GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE WORKING IN IT, HEALTHCARE AND AVIATION SECTOR

 PRODUCT CATEGORY BASED ANALYSIS

```python
ax = sns.countplot(data= df, x= 'Product_Category', hue='Product_Category', width= 0.7)
for bars in ax.containers:
    ax.bar_label(bars)
```

 we can see that most of the sold products are from food , clothing and electronics category
